Look like we need to do some data cleansing as layover flights have different seating arrangements bunched up together,
I believe the best approach would be to seperate all the seats into their own individual columns. 

```python
cabins = []
num_flights = []

#Interates through each individual entries of the 'CabinCode' column
for i in flight_df['CabinCode']:
    num_flights.append(i.count('||')+1)
    
    #Splits on the '||' characters to give us the individual seat classes
    cabins.append(i.split("||"))
    
#Create a new dataframe to store the seat class list and create the columns for the various seats
cabin_df = pd.DataFrame(cabins, columns=['FlightSeat1', 'FlightSeat2', 'FlightSeat3', 'FlightSeat4'])

num_flight_df = pd.DataFrame(num_flights, columns=['num_flight'])

cabin_df.fillna('N/A', inplace=True)
```

```python
flight_df.head()
```

```python
cabin_df.head()
```

We need to split each of the airplane types flown since they're bunched up together with the '||' char. I believe the best approach would be to have a column dedicated to each layover flight.

```python
#Initalize a list to keep seperated 
airplanes = []

#Iterates through all entries in the column
for i in flight_df['EquipmentDescription']:
    
    #Split if '||' is found
    airplanes.append(i.split("||"))

#Creates dataframe to store the splitted airplane types
airplane_df= pd.DataFrame(airplanes, columns=['Airplane1', 'Airplane2', 'Airplane3', 'Airplane4'])

#Fill empty entries with 'N/A' since the means layover flight is not applicable
airplane_df.fillna('N/A', inplace=True)
```

```python
#convert the total travel distance from string to float
flight_df['totalTravelDistance'] = flight_df['totalTravelDistance'].astype(float)
```

```python
#fill the empty entries with the mean of the entire travel distance column
flight_df['totalTravelDistance'].fillna(flight_df['totalTravelDistance'].mean() , inplace=True) 
```

```python
flight_df['totalTravelDistance'].value_counts()
```

```python
#convert total travel distance back to string now that we calculated the mean
flight_df['totalTravelDistance'] = flight_df['totalTravelDistance'].astype('string')
```

```python
#fill in any distance with empty entries with the total travel distance
flight_df.loc[flight_df['Distance'].isnull(),'Distance'] = flight_df['totalTravelDistance']
```

```python
#Flights with layovers have the distances of each individual flight in one column, we'll need to seperated them
dists = []

#Iterates through all of the entries in the distance column
for i in flight_df['Distance']:
    
    #If '||' is found seperate the distances and append them to the list
    dists.append(i.split("||"))

#Create a dataframe for the list holding the now split distances
dists_df= pd.DataFrame(dists, columns=['Dist1', 'Dist2', 'Dist3', 'Dist4'])

#Fill in any empty entries with 'N/A' as not all flights will have lay overs
dists_df.fillna('N/A', inplace=True)
```

```python
#Lets create a new dataframe with all of the new dataframes we've created included
new_flight_df = pd.concat([flight_df, cabin_df, airplane_df, dists_df, num_flight_df], axis=1)
```

```python
#Adds up the revenue generated by basefare and totalfare and create dataframes out of them
revenue_base = pd.DataFrame(new_flight_df.groupby('flightDate')['baseFare'].sum())
revenue_total = pd.DataFrame(new_flight_df.groupby('flightDate')['totalFare'].sum())
revenue = pd.merge(revenue_base, revenue_total, on='flightDate')
```

Lets visualize the sum of the fares for each day

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, sharey=True, figsize=(10,3))

#plotting base fare
sns.barplot(revenue, x=revenue.index, y="baseFare", ax=ax1)

#plotting total fare
sns.barplot(revenue, x=revenue.index, y="totalFare", ax=ax2)

#make x label look neater
ax1.tick_params(axis='x', labelrotation=90)
ax2.tick_params(axis='x', labelrotation=90)

#lets try a horzontal line at the 6 million dollar mark
ax1.axhline(y=6000000, c='purple')
ax2.axhline(y=6000000, c='purple') 

plt.tight_layout()
```

As expected totalfare has a higher overall incoming for everyday because of the added costs. The income seems to be stronger in the first half on the plot and then it drops off quite significantly after the 26th, what could be the reason for this? Lets add up the total incoming comparing the two different fares.

```python
#Adds up all the ticket income for basefare
base_sum = round(revenue['baseFare'].sum(), 2)

#Adds up all the ticket income for totalfare
total_sum = round(revenue['totalFare'].sum(), 2)
```

```python
#Plot the sum of basefare and totalfare for comparsion 
sns.set()
revenue.T.plot(kind='bar', stacked=True)

#Put the legend outside of the plot
plt.legend(bbox_to_anchor=(1.25, 1), borderaxespad=0)

#Disable scientific notation on the y-axis
plt.gcf().axes[0].yaxis.get_major_formatter().set_scientific(False)

#Print the total amount made during the timeframe on the x-axis
plt.text(0, 0, base_sum, horizontalalignment='center',verticalalignment='top')
plt.text(1, 0, total_sum, horizontalalignment='center',verticalalignment='top')
```

Base fare made a total of \\$75,459,745.44 throughout the 19-day timeframe, total fare made a total of \\$87,400,858.69. So there is a difference of \\$11,941,113 or a 15.82% increase.
